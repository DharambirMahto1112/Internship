2. In the above question, now scrape the following details of each product listed in first 3 pages
of your search results and save it in a data frame and csv. In case if any product vertical has
less than 3 pages in search results then scrape all the products available under that product
vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details"
and “Product URL”. In case, if any of the details are missing for any of the product then
replace it by “-“

```python
href = []
pages = list(range(1,4))
for page in pages:
    url = 'https://www.amazon.in/s?k=samsung&page={}&qid=1634754658&ref=sr_pg_{}'.format(page,page)
    lnks1=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    for lnk in lnks1:
        href.append(lnk.get_attribute('href'))
```

```python
print(len(href))
```

```python
Brand_Name = []
Name_Product = []
Rating = []
No_of_Ratings = []
Price = []
Return_Exchange = []
Expected_Delivery = []
Availability = []
Other_Details = []
Product_URL = []
```

```python
for i in href:
    driver.get(i)

    # extracting Brand names
    try:
        brand = driver.find_element_by_xpath("//*[@id='productOverview_feature_div']/div/table/tbody/tr[3]/td[2]/span")
        Brand_Name.append(brand.text)
    except NoSuchElementException:
        Brand_Name.append('NaN')
    
    # extracting product name
    try:
        pn = driver.find_element_by_xpath("//*[@id='productOverview_feature_div']/div/table/tbody/tr[1]/td[2]/span")
        Name_Product.append(pn.text)
    except NoSuchElementException:
        Name_Product.append('NaN')
    
    # extracting rating of the product
    try:
        rating = driver.find_element_by_xpath("//*[@id='reviewsMedley']/div/div[1]/div[2]/div[1]/div/div[2]/div/span/span")
        Rating.append(rating.text)
    except NoSuchElementException:
        Rating.append('NaN')
    
    # extracting no. of rating
    try:
        nr = driver.find_element_by_xpath("//*[@id='reviewsMedley']/div/div[1]/div[2]/div[2]/span")
        No_of_Ratings.append(nr.text)
    except NoSuchElementException:
        No_of_Ratings.append('NaN')
    
    # extracting price of the product
    try:
        price = driver.find_element_by_xpath("//td[@class='a-span12']")
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('NaN')
    
    # extracting return and exchange of the product
    try:
        re = driver.find_element_by_xpath("//*[@id='RETURNS_POLICY']/span/div[2]/a")
        Return_Exchange.append(re.text)
    except NoSuchElementException:
        Return_Exchange.append('NaN')
    
    # extracting expected delivery
    try:
        ed = driver.find_element_by_xpath("//*[@id='ddmDeliveryMessage']/b")
        Expected_Delivery.append(ed.text)
    except NoSuchElementException:
        Expected_Delivery.append('NaN')
    
    # extracting availability
    try:
        av = driver.find_element_by_xpath("//*[@id='availability']/span")
        Availability.append(av.text)
    except NoSuchElementException:
        Availability.append('NaN')
 
    # extracting other details
    try:
        od = driver.find_element_by_xpath("//*[@id='feature-bullets']")
        Other_Details.append(od.text)
    except NoSuchElementException:
        Other_Details.append('NaN')

# product url
Product_URL = href
```

```python
df = pd.DataFrame({})
df['Brand_Name'] = Brand_Name 
df['Name of the Product'] = Name_Product
df['Rating'] = Rating
df['No_of_Ratings'] = No_of_Ratings
df['Price'] = Price
df['Return_Exchange'] = Return_Exchange
df['Expected_Delivery'] = Expected_Delivery
df['Availability'] = Availability
df['Other_Details'] = Other_Details
df['Product_URL'] = Product_URL 
df
```

```python
# Requesting Chrome driver to access the link:
driver = webdriver.Chrome(r"C:\Program Files (x86)\chromedriver.exe")
driver.maximize_window()
driver.get("https://www.google.co.in/imghp?hl=en&authuser=0&ogbl")

# inserting data as per the required search  key:
search = driver.find_element_by_xpath("//input[@class='gLFyf gsfi']")
search.send_keys('fruits')

# clicking on search:
serch_btn = driver.find_element_by_xpath("//button[@class='Tg7LZd']").click()


# auto scroll
page_scroll_sleep = 2

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(page_scroll_sleep)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")

    if new_height == last_height:
    #break #insert press load more
        try:
            element = driver.find_elements_by_class_name('mye4qd') #returns list
            element[0].click()
        except:
            break
    last_height = new_height
    

# getting the urls of images
src = []
lnks1=driver.find_elements_by_xpath("//img[@class='rg_i Q4LuWd']")
for lnk in lnks1[0:100]:
    src.append(lnk.get_attribute('src'))
    

# saving the images from url
from urllib.request import urlopen
for index, url in enumerate(src):
    link = urlopen(url)
    try:
        name = "fruits%s.jpg" % (index+1)
        with open(name, "wb") as output:
            output.write(link.read())
    except IOError:
        print ("Unable to create %s"% name)
```
