```python
driver = webdriver.Chrome(r"C:\Program Files (x86)\chromedriver.exe")
driver.get("https://www.glassdoor.co.in/Salaries/index.htm")
```

```python
search_job = driver.find_element_by_id("KeywordSearch")
search_job.send_keys("Data Scientist")
search_loc = driver.find_element_by_xpath("//input[@id ='LocationSearch']")
search_loc.send_keys('Noida')
serch_btn = driver.find_element_by_xpath("//button[@id ='HeroSearchButton']")
serch_btn.click()
```

```python
Company_name = []
no_of_salary = []
avg_salary = []
min_salary = []
max_salary = []
demo_min = []

c_name = driver.find_elements_by_xpath("//a[@class = 'css-f3vw95 e1aj7ssy3']")
AS = driver.find_elements_by_xpath("//div[@class = 'col-12 col-lg-4 px-lg-0 d-flex align-items-baseline']")
ns = driver.find_elements_by_xpath("//div[@class = 'col-12 col-lg-auto']")
min_s = driver.find_elements_by_xpath("//div[@class='d-flex mt-xxsm css-79elbk epuxyqn0']/p[@class='m-0 css-1vkj9it']")

for i in c_name[0:10]:
    Company_name.append(i.text)
for i in AS[0:10]:
    avg_salary.append(i.text.replace('\n',''))
for i in ns[0:10]:
    no_of_salary.append(i.text)
for i in min_s[0:20]:
    demo_min.append(i.text)
for i in range(0,len(demo_min)):
    if i%2:
        max_salary.append(demo_min[i])
    else:
        min_salary.append(demo_min[i])



df = pd.DataFrame({'Company Name':Company_name,
                    'Number of Salary':no_of_salary,
                    'Average Salary':avg_salary,
                    'Minimum salary':min_salary,
                    'Maximum Salary':max_salary})
df
```

Q6 : Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:

```python
driver = webdriver.Chrome(r"C:\Program Files (x86)\chromedriver.exe")
driver.get("https://www.flipkart.com/")
```

```python
search_item = driver.find_element_by_xpath("//input[@class ='_3704LK']")
search_item.send_keys("sunglasses")
serch_btn = driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
serch_btn.click()
```

```python
Brand = []
Prodt_desc = []
Price = []
off = []
href = []
```

```python
pages = list(range(1,4))
for page in pages:
    url = 'https://www.flipkart.com/search?q=sunglasses&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&page={}'.format(page)
    lnks1=driver.find_elements_by_xpath("//a[@class='IRpwTa']")
    for lnk in lnks1:
        href.append(lnk.get_attribute('href'))
```

```python
href = href[0:100]
print(len(href))
```

```python
for i in href:
    driver.get(i)

    # extracting Brand names
    try:
        brand = driver.find_elements_by_xpath("//span[@class='G6XhRU']")
        for j in brand:
            Brand.append(j.text)
    except NoSuchElementException:
            Brand.append('NaN')

    # extracting price
    try:
        rate = driver.find_elements_by_xpath("//div[@class='_30jeq3 _16Jk6d']")
        for i in rate:
            Price.append(i.text)
        
    except NoSuchElementException:
            Price.append('NaN')

    # extracting description
    try:
        desc = driver.find_elements_by_xpath("//h1[@class='yhB1nd']/span[2]")
        for l in desc:
            Prodt_desc.append(l.text)
    except NoSuchElementException:
            Prodt_desc.append('NaN')

    # extracting discount
    try:
        lnks=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb _31Dcoz pZkvcx']")
        for i in lnks:
            off.append(i.text)
    except NoSuchElementException:
        off.append('NaN')
        
```

```python
print(len(Brand))
print(len(Price))
print(len(Prodt_desc))
print(len(off))
```

```python
df = pd.DataFrame({'Brand Name':Brand,
                  'Price':Price,
                  'Product Description':Prodt_desc,
                  'Discount':off})
df
```

```python
driver.close()
```
